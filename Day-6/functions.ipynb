```python
# factors
def find_factors(n):
    print("Factors of ",n,"are:")

    for i in range(1, n+1):
        if n % i == 0:
            print(i)

# calling the function
num = int(input("Enter a number: "))
find_factors(num)     
```

```python
''' Create a function that check if username and password is correct update IS_Admin global variable '''

isAdmin = False
def update():
    global isAdmin
    password = "123456789"
    if(password == "123456789"):
        isAdmin = True
        return isAdmin

a = update()
print(a)
```

```python
''' **kwargs --> many keywords'''

user_profile = {}
def profile(**data):
    global user_profile
    print(type(data))
    print(data)

profile(name = 'Tom' , age = '54')
    
def update(**data):
    global user_profile
    
    print(data)

update(sem = '6th')
```

# lambda

```python
add = lambda a,b:a+b
dif = lambda a,b:a-b
mul = lambda a,b:a*b
div = lambda a,b:a/b
print(add(2,3))
print(dif(9,3))
print(mul(2,3))
print(div(2,3))
```

```python
''' lamda with condition '''
max_val = lambda a,b: a if  a > b else b
max_val (1,2)
```

```python
eo = lambda a,b : "Even" if a%b==0 else "odd"
eo(9,3)
```

```python
def outer_fun():
    print("outer function")
    def inner_fun():
        print("i am inner funtion")

    inner_fun()

outer_fun()
```
