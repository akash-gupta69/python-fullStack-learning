```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def say_hi():
    print("Hi")

# say_hi = my_decorator(say_hi)
```

```python
def decorator(func):
    def wrapper(name):
        print("Hello starts")
        func(name)
        print("Hello ends")
    return wrapper
    
 
@decorator
def greet(name):
    print(f"Hi {name}")
```

```python
def login_required(func):
    def wrapper():
        print("Checking login...")
        func()
    return wrapper
    

@login_required
def dashboard():
    print("Welcome to dashboard")
```
